La base de datos a estudiar es **Monthly Electricity Production in GWh [2010-2022]**, se encuentra en el siguiente enlace: https://www.kaggle.com/datasets/ccanb23/iea-monthly-electricity-statistics

El dataset se recopiló del sitio web de International Energy Agency (IEA), incluye información de la producción de energía mensual de 2010 a 2022. La producción de energía es medida en GigaWatt-hora (Gwh) y está clasificada en los siguientes tipos: hydro, wind, solar, geothermal, nuclear, fossil fuels, y otras.

Los países presentes en la base de datos son:
Argentina, Australia, Austria, Belgium, Brazil, Bulgaria, Canada, Chile, Colombia, Costa Rica, Croatia, Cyprus, Czech Republic, Denmark, Estonia, Finland, France, Germany, Greece, Hungary, IEA Total, Iceland, India, Ireland, Italy, Japan, Korea, Latvia, Lithuania, Luxembourg, Malta, Mexico, Netherlands, New Zealand, North Macedonia, Norway, OECD Americas, OECD Asia Oceania, OECD Europe, OECD Total, People's Republic of China, Poland, Portugal, Republic of Turkiye, Romania, Serbia, Slovak Republic, Slovenia, Spain, Sweden, Switzerland, United Kingdom, United States.

Los tipos de electricidad en la base de datos son:
* Hydro: Producción neta de instalaciones hidroeléctricas.
* Wind: Electricidad producida en generadores heólicos terrestres y marinos.
* Solar: Electricidad producida usando radiación solar. Incluye energía fotovoltaica y térmica.
* Geothermal: Electricidad generada a partir del calor emitido desde el interior de la corteza terrestre, generalmente en forma de agua caliente o vapor.
* Other renewables: Electricidad generada a partir de mareas, olas, océanos y otras fuentes no combustibles.
* Nuclear: Electricidad producida a partir del calor generado por la fisión nuclear.
* Total combustible fuels: Producción a partir de combustibles fósiles (carbón primario, productos de carbón, turba y productos de turba, esquisto bituminoso y arenas bituminosas, petróleo crudo, LGN, productos derivados del petróleo, gas natural) y combustibles renovables y residuos (biocombustibles sólidos, biogases, biocombustibles líquidos)
* Coal: Producción a partir de carbón primario
* Oil: Producción a partir de carbón primario, productos de carbón, turba y productos de turba, esquisto bituminoso y arenas bituminosas, y gas manufacturado
* Natural gas: Producción a partir de petróleo crudo, líquidos de gas natural, materias primas de refinería y productos derivados del petróleo. 
* Combustible renewables: Producción de combustibles renovables (biocombustibles sólidos, biogases, biocombustibles líquidos)
* Other combustible non-renewables
* Not specified 
* Net electricity production: Suma de producción neta de producción de electricidad autóctona por fuente de energía
* Total imports
* Total exports
* Electricity supplied: Producción neta de electricidad + Importaciones - Exportaciones. Incluye pérdidas de Transmisión y Distribución.
* Used for pumped storage: Electricidad consumida por bombeo de agua a reserva en centrales hidroeléctricas mixtas y puras de bombeo.
* Distribution losses: Pérdidad debidas al transporte y distribución de energía eléctrica
* Final consumption: Eleectricidad disponible (producción neta + importaciones - exportaciones - uso para bombeo - pérdidas)
* Electricity trade: Comercio de electricidad
* Renewables: Incluye la generación de electricidad a partir de energía hidráulica, eólica, solar, geotérmica, otras renovables y renovables combustibles.
* Non-renewables: Incluye generación eléctrica a partir de Carbón, Petróleo, Gas Natural, Otros Combustibles, Nuclear y Fuentes no especificadas.
* Others
* Other renewables aggregated
* Low carbon: Renovables y generación nuclear de energía
* Fossil fuels: Combustibles fósiles

Las columnas de la base de datos son:
* COUNTRY: Nombre del país
* CODE_TIME: Código para especificar el mes y el año (e.g., JAN2010 para January 2010)
* TIME: Mes y año (e.g., January 2010)
* YEAR: Año del dato
* MONTH: Mes del dato en formato (1-12)
* MONTH_NAME: TMes del dato como cadena (e.g., January)
* PRODUCT: Tipo de energía producida (e.g., Hydro, Wind, Solar)
* VALUE: Cantidad de electricidad producida en gigawatt-hora (GWh)
* DISPLAY_ORDER: Orden en que los datos deben ser desplegados
* yearToDate: Cantidad de electricidad generada para el año correspondiente hasta el año correspondiente en GWh 
* previousYearToDate: Cantidad de electricidad generada para el año anterior hasta el mes correspondiente en GWh
* share: La contribución del tipo de energía en la generación total de electricidad para el país en formato decimal

Esta base de datos resulta muy interesante, pues estudiándola se pueden deducir hechos en la producción de electricidad, por ejemplo, los países que producen más y menos electricidad o el método de producción más y menos utilizado. Esta información puede utilizarse para tomar medidad para reducir la contaminación.

Por otro lado, también es importante estudiar el efecto de la pandemia en la producción de electricidad.

Como puede observarse, algunas categorías en los tipos de electricidad son combinaciones de otras, por lo que en este trabajo solo se tomaran en cuenta las categorías:  **Hydro, Wind, Solar, Geothermal, Other renewables, Nuclear, Coal, Oil, Natural gas, Others y Fossil fuels.**

De manera similar, en la coumna de country se incluyen organizaciones conformadas por varios países. Por lo anterior, solo se consideraran países individuales.

```python
#Importamos las librerías necesarias
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Leemos la base de datos
data =  pd.read_csv('ElectricityProduction.csv')
```

```python
data
```

```python
#En la columna PRODUCT, seleccionamos las cateogorías de interés
data = data[(data.PRODUCT.isin(['Hydro', 'Wind', 'Solar','Geothermal', 'Other renewables', 'Nuclear', 'Coal', 'Oil', 'Natural gas', 'Others', 'Fossil fuels']))]
data
```

```python
#En la columna COUNTRY, eliminamos a las organizaciones conformadas por varios países
data = data[~(data.COUNTRY.isin(['IEA Total', 'OECD Americas', 'OECD Asia Oceania', 'OECD Europe', 'OECD Total']))]
data
```

```python
#Ejecutamos un análisis rápido de la base de datos
data.describe(include='all').T
```

De la tabla anterior se observa que:
* Francia es el país del que se tienen más registros
* La fuente de producción de electricidad de la que se tienen más registros es **Combustibles Fósiles** hasta el año 2022.
* El més y el año del que más se tienen registros es Mayo de 2021, en plena pandemia.
* En promedio se producen 312.662846 GWh por mes.
* La máxima cantidad de electricidad que se ha producido en un mes es 304600.315 GWh

###  <center> <font color = 'orchid'> 5 países que más han producido electricidad entre 2010 y 2022 y 5 países que menos han producido electricidad entre 2010 y 2020

```python
#Agrupamos la columna 'VALUE' por la columna 'COUNTRY' y sumamos su producción total de electricidad
paisesval = data.groupby(by='COUNTRY')['VALUE'].sum()
#Ordenamos de menor a mayor
paisesval = paisesval.sort_values().to_frame()
```

```python
#Imprimimos los 5 primeros lugares en la lista
paisesval[:5]
```

El resultado anterior es razonable, pues los países que menos han producido electricidad entre 2010 y 2020 tienen una extensión territorial reducida.

Los cinco países que **más han producido energía entre los años 2010 y 2022** son:

```python
#Imprimimos los últimos cinco lugares en la lista
paisesval[-5:]
```

El país que más ha producido electricidad entre los años 2010 y 2020 es USA. Japón, India, Corea y Alemania son los otros países que más han producido electricidad en este periodo. Este resultado constrasta la hipótesis de que la producción de electricidad es proporcional a la extensión territorial del país por la presencia de Japón y Corea.

USA, el primer lugar, produce alrededor de 3 veces más electricidad que Japón, el segundo lugar. La producción de electricidad de USA es aproximadamente 3552 veces la producción de Costa Rica. 

###  <center> <font color = 'orchid'> Método de producción de electricidad más y menos utilizado

```python
#Agrupamos la columna 'VALUE' por la columna 'PRODUCT' y sumamos su producción total de electricidad
prodval = data.groupby(by='PRODUCT')['VALUE'].sum()
#Ordenamos de menor a mayor
prodval = prodval.sort_values().to_frame()
prodval
```

Como era de esperarse, el método de producción más utilizado es **Combustibles Fósiles**, seguido por **Geotérmica**. El método de producción menos utilizado es **Otras renovables**, es decir, a partir de mareas y olas. Se observa que al sumar las contribuciones de **Solar, Wind, y Hydro** se obtiene un valor cercano a la producción de combustibles fósiles. Por lo que se deben aumentar esfuerzoz en estos métodos para aumentar la producción y reducir el uso de combustibles fósiles. Esto se puede apreciar mejor en la siguiente gráfica

###  <center> <font color = 'orchid'> Variación de la producción de electricidad por año

```python
sns.kdeplot(data=data, x="VALUE", hue="YEAR", palette='Spectral').set(title='Distribución de la producción \n mensual de electricidad por año')
```

De la gráfica anterior se observa que la producción mensual de electricidad ha ido aumentando drásticamente. La lantidad que más se produce del año 2020 es más del doble de la cantidad que más se produce del año 2010, este año corresponde a la pandemia por COVID-19. Del año 2015 al año 2016 hubo un aumento considerable respecto a los años anteriores. En el año 2018 también hubo un aumento importante. En todos los años se observa el sesgo de grandes cantidades de electricidad producidas en algunos meses.



###  <center> <font color = 'orchid'> Variación de la producción de electricidad por mes

```python
sns.kdeplot(data=data, x="VALUE", hue="MONTH_NAME", palette='Paired').set(title='Distribución de la producción \n mensual de electricidad por mes')
```

Se observa que durante el mes de **Abril** la producción mensual de electricidad alcanza un pico más alto que los otros mese, mientras que en **Julio** la moda de la producción mensual de electricidad es menor. Sin embargo, los valores mayores a 250 000 GWh se presentan en los meses de **julio, junio y enero**, esto puede deberse al clima y al uso de aires acondicionados o calefactores.
